```python
#Import Dependencies
import gym
import highway_env
import numpy as np
from stable_baselines import HER, SAC, PPO2, DQN
```

```python
#Roundabout
env = gym.make('roundabout-v0')
model = PPO2('MlpPolicy', env, verbose=1)
```

```python
#Random Action
for i in range(10):
    done = False
    env.reset()
    while not done:
        env.render()
        action = env.action_space.sample()
        next_state, reward, done, info = env.step(action)
        print(info)
        print(done)
        
env.close()
```

```python
#Create Model
model = PPO2('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=10000)
```

```python
#Save & load model

model.save('roundabout')
del model
```

```python
model = PPO2.load('roundabout')
```

```python
#Visualize Model
for i in range(10):
    done = False
    obs = env.reset()
    while not done:
        env.render()
        action, _states = model.predict(obs)
        next_state, reward, done, info = env.step(action)
        print(info)
        print(done)
        
env.close()
```
